```python
import re

def validate_email(email):
    # Улучшенный паттерн: допускает домены длиной 2+ символа
    pattern = r'^[\w\.-]+@[\w\.-]+\.[a-zA-Z]{2,}$'
    return bool(re.match(pattern, email))

# Тесты
print(validate_email('example@google.com'))      # True
print(validate_email('invalid@email'))           # False
print(validate_email('another@yandex.ru.co'))    # True
print(validate_email('test@domain.museum'))      # True
```

```python
import re

def extract_phone_numbers(text):
    # Паттерн учитывает различные форматы
    pattern = r'\(?\d{3}\)?[-\s]?\d{3}[-\s]?\d{4}\b'
    return re.findall(pattern, text)

text = "Свяжитесь со мной по номерам (967) 060-2424 или (495) 987-6543. Мой старый номер был 555-123-4567."
phone_numbers = extract_phone_numbers(text)
print(phone_numbers)
# Результат: ['(967) 060-2424', '(495) 987-6543', '555-123-4567']
```

```python
import re

def remove_html_tags(text):
    # Паттерн удаляет любые теги, включая самозакрывающиеся
    pattern = r'<[^>]*>'
    return re.sub(pattern, '', text)

html_text = "<p>Это <b>пример</b> с HTML-тегами.</p><br/>"
plain_text = remove_html_tags(html_text)
print(plain_text)
# Результат: Это пример с HTML-тегами.
```

```python
import re

def censor_words(text, words_to_censor):
    # Паттерн с границами слов (\b) для точного совпадения
    pattern = r'\b(' + '|'.join(re.escape(word) for word in words_to_censor) + r')\b'
    return re.sub(pattern, lambda match: '*' * len(match.group()), text, flags=re.IGNORECASE)

text = "Это предложение содержит запрещенные слова: Кошка, Собака и Мышка."
censored_text = censor_words(text, ['Кошка', 'Собака', 'Мышка'])
print(censored_text)
# Результат: Это предложение содержит запрещенные слова: *****, ****** и ******.
```

```python
import re
from datetime import datetime

def validate_user_input(name, email, phone, birth_date):
    # Имя: только буквы и пробелы (кириллица + латиница)
    name_pattern = r'^[a-zA-Zа-яА-ЯёЁ\s]+$'
    email_pattern = r'^[\w\.-]+@[\w\.-]+\.[a-zA-Z]{2,}$'
    phone_pattern = r'^\d{3}-\d{3}-\d{4}$'

    if not re.match(name_pattern, name):
        return "Некорректное имя"
    if not re.match(email_pattern, email):
        return "Некорректный адрес электронной почты"
    if not re.match(phone_pattern, phone):
        return "Некорректный номер телефона"

    # Проверка корректности даты
    try:
        datetime.strptime(birth_date, '%d/%m/%Y')
    except ValueError:
        return "Некорректная дата рождения"

    return "Данные введены корректно"

# Тест
name = "Иван Петров"
email = "ivan.petrov@example.com"
phone = "123-456-7890"
birth_date = "01/01/1990"

validation_result = validate_user_input(name, email, phone, birth_date)
print(validation_result)
# Результат: Данные введены корректно
```

```python
import re

def detect_bad_words(text, bad_words):
    # Паттерн с границами слов для точного совпадения
    bad_words_pattern = r'\b(' + '|'.join(re.escape(word) for word in bad_words) + r')\b'
    matches = re.findall(bad_words_pattern, text, re.IGNORECASE)
    return matches

text = "Это предложение содержит плохие слова: невоспитанный жадина."
bad_words = ["Жадина", "Невоспитанный", "Руки-крюки"]

detected_bad_words = detect_bad_words(text, bad_words)

if detected_bad_words:
    print(f"В тексте обнаружены нежелательные слова: {', '.join(detected_bad_words)}")
else:
    print("В тексте нет нежелательных слов")
# Результат: В тексте обнаружены нежелательные слова: невоспитанный, жадина
```
